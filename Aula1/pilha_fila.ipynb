```python
class pilha:
    #construtor que deve receber o tamanho da pilha
    def __init__(self,tam):
        self.item=[]
        self.tam=tam
    
    #funcao para inserir elemento
    def push(self,elemento):
        x=self.is_full()
        if x==True:
            print "Pilha cheia"
            return
        else:
            self.item.append(elemento)
            
    #funcao para remover elemento            
    def pop(self):
        y=self.is_empty()
        if y==True:
            print "Pilha vazia"
            return
        else:
            print self.item.pop()            
            
            
    #funcao para checar se a pilha esta vazia
    def is_empty(self):
        if len(self.item)==0:
            return True
        else:
            return False
            
    #funcao para checar se a pilha esta cheia
    def is_full(self):
        if len(self.item)==self.tam:
            return True
        else:
            return False
            

        
    #funcao para ver primerio item da pilha    
    def peek(self):
        print self.item[len(self.item)-1]
```

```python
class fila:
    #construtor que deve receber o tamanho da fila
    def __init__(self,tam):
        self.item=[]
        self.tam=tam
        
    #funcao para ver primerio item da fila 
    def peek(self):
        print self.item[0]
        
    #funcao para checar se a fila esta vazia    
    def is_empty(self):
        if len(self.item)==0:
            return True
        else:
            return False
        
    #funcao para checar se a fila esta cheia
    def is_full(self):
        if len(self.item)==self.tam:
            return True
        else:
            return False
        
    #funcao para inserir elemento    
    def enqueue(self,elemento):
        x=self.is_full()
        if x==True:
            print "Fila cheia"
            return
        else:
            self.item.append(elemento)
            
    #funcao para remover elemento      
    def dequeue(self):
        y=self.is_empty()
        if y==True:
            print "Fila vazia"
            return
        print self.item.pop(0)
```
