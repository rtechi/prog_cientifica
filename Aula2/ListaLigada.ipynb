```python
class fila(ListaLigada):


# Classe Fila
# Fila implementada em python

# Funcoes:
# remove o primeiro elemento da fila (dequeue)
# insere um elemento da fila (enqueue)
# apresenta o primeiro elemento da fila (first)




    def enqueue(self,valor):
        self.insere_tras(valor)

    def dequeue(self):
        valor = self.remove_frente()
        return valor

    def first(self):
        return self.get_start()
```

```python
class pilha(ListaLigada):
    
    
#     Classe Pilha
#     Implementação da Estrutura de Dados Pilha

#     Com funções de:
#     -- Inicialização de variáveis (__init__)
#     -- Retirada de uma valor do topo da Pilha (pop)
#     -- Inserção de um valor no topo da Pilha (push)
#     -- Apresentação do valor do topo da Pilha (top)


    def push(self,valor):
        self.insere_frente(valor)

    def pop(self):
        try:
            fim = self.remove_frente()
            return fim
        except IndexError:
            return None

    def top(self):
        return self.get_end()
```
