```python
class No():
    
#     Classe No
#     No implementado em python

#     Funcoes:
#     inicialização de variáveis (__init__)
#     altera valor do nó (set_valor)
#     altera valor apontado (set_pointer)
#     retorna valor do nó (get_valor)
#     retorna valor apontado  (get_pointer)


    def __init__(self,valor,pointer=None):
        self.valor=valor
        self.pointer=pointer
    
    def set_valor(self,valor):
        self.valor=valor
    
    def set_pointer(self,valor):
        self.pointer=valor
    
    def get_valor(self):
        return self.valor
    
    def get_pointer(self):
        return self.pointer
    
    def __del__(self):
        self.valor = None
        self.proximo = None
        #print ""
        
```

```python
class fila(ListaLigada):


# Classe Fila
# Fila implementada em python

# Funcoes:
# remove o primeiro elemento da fila (dequeue)
# insere um elemento da fila (enqueue)
# apresenta o primeiro elemento da fila (first)




    def enqueue(self,valor):
        self.insere_tras(valor)

    def dequeue(self):
        valor = self.remove_frente()
        return valor

    def first(self):
        return self.get_start()
```

```python
class pilha(ListaLigada):
    
    
#     Classe Pilha
#     Implementação da Estrutura de Dados Pilha

#     Com funções de:
#     -- Inicialização de variáveis (__init__)
#     -- Retirada de uma valor do topo da Pilha (pop)
#     -- Inserção de um valor no topo da Pilha (push)
#     -- Apresentação do valor do topo da Pilha (top)


    def push(self,valor):
        self.insere_frente(valor)

    def pop(self):
        try:
            fim = self.remove_tras()
            return fim
        except IndexError:
            return None

    def top(self):
        return self.get_end()

```

```python
class DFS(pilha,Jogo):
    def __init__(self):
        self.objetivo=[1,2,3,4,5,6,7,8,0]
        self.em_uso=[0,0,0,0,0,0,0,0,0]
        self.em_uso_ancora=[[0,0,0,0,0,0,0,0,0]]
        self.pos_0_dfs=0
        self.ja_visto=[]
        self.lista_tabelas=pilha()
        self.lista_tabelas.push([4,1,6,3,2,8,7,0,5])
        self.num_=0
        
        
    def solve(self):
        start = time.time()
        #checa se tabela em uso é o objetivo
        while self.em_uso!=self.objetivo:
            #seleciona tabela a ser analisada
            self.em_uso=self.lista_tabelas.top()
            #verifica se tabela em uso ja foi vista
            a=next((True for elem in self.ja_visto if array_equal(elem, self.em_uso)), False)
            if a==True:
                #remove elemento ja visto da lista de tabelas
                self.lista_tabelas.pop()
                continue
            else:
                #descobre os movimentos possiveis da tabela atual
                self.movimentos_possiveis=self.tabela_de_movimentos(self.get_posicao(self.em_uso))
                #coloca o valor a em uso no ja visto
                self.ja_visto.append(self.lista_tabelas.top())
                #atualiza a lista de tabelas
                for x in range(len(self.movimentos_possiveis)):
                    self.lista_tabelas.push(self.movimenta(self.movimentos_possiveis[x],self.em_uso))
                #remove tabela que estava sendo analisada
                self.lista_tabelas.pop()
                self.num_=self.num_+1
        end = time.time()
        print ("Resultado DFS",self.em_uso,(end - start))   
```

```python
class BFS(fila,Jogo):
    def __init__(self):
        self.objetivo=[1,2,3,4,5,6,7,8,0]
        self.inicio=[4,1,6,3,2,8,7,0,5]
        self.em_uso=[0,0,0,0,0,0,0,0,0]
        self.em_uso_ancora=[[0,0,0,0,0,0,0,0,0]]
        self.pos_0_dfs=0
        self.ja_visto=[]
        self.lista_tabelas=fila()
        self.lista_tabelas.enqueue([4,1,6,3,2,8,7,0,5])
        self.num_=0
        
        
    def solve(self):
        start = time.time()
        while self.em_uso!=self.objetivo:
            #seleciona tabela a ser analisada
            self.em_uso=self.lista_tabelas.first()
            a=next((True for elem in self.ja_visto if array_equal(elem, self.em_uso)), False)
            if a==True:
                #remove ja visto da lista 
                self.lista_tabelas.dequeue()
                continue
            else:
                #pega os movimentos possiveis
                self.movimentos_possiveis=self.tabela_de_movimentos(self.get_posicao(self.em_uso))
                #coloca o valor a ser analisado no ja visto
                self.ja_visto.append(self.lista_tabelas.first())
                #atualiza tabela
                for x in range(len(self.movimentos_possiveis)):
                    self.lista_tabelas.enqueue(self.movimenta(self.movimentos_possiveis[x],self.em_uso))    
                #remove valor a ser analisado da lista
                self.lista_tabelas.dequeue()
                self.num_=self.num_+1
        end = time.time()
        print ("Resultado BFS",self.em_uso,(end - start))
```
